# Lab Session 9: Limits of Diversification

```python
import pandas as pd
import edhec_risk_kit as erk
import numpy as np

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
ind_return = erk.get_ind_returns()
ind_size = erk.get_ind_size()
ind_nfirms = erk.get_ind_nfirms()
```

```python
print(ind_return.shape,
      ind_size.shape,
      ind_nfirms.shape)
```

```python
ind_nfirms.head()
```

```python
ind_mktcap = ind_nfirms * ind_size
ind_mktcap.shape
```

```python
total_mktcap = ind_mktcap.sum(axis="columns")
total_mktcap.plot(figsize=(10,5))
```

```python
ind_capweight = ind_mktcap.divide(total_mktcap, axis="rows")
ind_capweight.head()
```

```python
ind_capweight["1926"].sum(axis="columns")
```

```python
ind_capweight[["Fin","Steel"]].plot(figsize=(10,5))
```

```python
total_market_return = (ind_capweight*ind_return).sum(axis="columns")
total_market_return.plot(figsize=(10,5))
```

```python
total_market_index = erk.drawdown(total_market_return).Wealth
total_market_index.plot(figsize=(10,5),title="Total Markeet CapWeighted Index 1926-2018")
```

```python
total_market_index["1980":].plot(figsize=(10,5))
total_market_index["1980":].rolling(window=36).mean().plot(figsize=(10,5))
# Add 36 month moving average
```

```python
tmi_tr36rets = total_market_return.rolling(window=36).aggregate(erk.annualize_rets, periods_per_year=12)
total_market_return.plot(label="Returns",legend=True, color="k",linewidth=0.5)
tmi_tr36rets.plot(figsize=(10,5), label="Tr36 mo Return",legend=True,color="red")
```

### Rolling Correlation - along with MultiIndices and `.groupby`

```python
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.tail()
```

```python
ind_return.head()
```

```python
ts_corr.index.names = ['date','industry']
ts_corr.tail()
```

```python
ind_tr36corr = ts_corr.groupby(level='date').apply(lambda corrmat: corrmat.values.mean())
ind_tr36corr.plot(figsize=(10,5))
```

```python
tmi_tr36rets.plot(label="Tr36 Mo Rets",figsize=(10,5),legend=True)
ind_tr36corr.plot(label="Tr36 Mo Corr",legend=True,secondary_y=True)
```

```python
tmi_tr36rets["2006":].plot(label="Tr36 Mo Rets",figsize=(10,5),legend=True)
ind_tr36corr["2006":].plot(label="Tr36 Mo Corr",legend=True,secondary_y=True)
```
